```python
# 1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.
'''
Given : T(0) = 5

PUT n = 1
T(1) = 3*T(1-1) + 12*1
T(1) = 3*T(0) + 12
T(1) = 3*5 + 12
T(1) = 27

PUT n = 2
T(2) = 3*T(2-1) + 12*2
T(2) = 3*T(1) + 24
T(2) = 3*27 + 24
T(2) = 81 + 24

T(2) = 105

'''
```

```python
# 2. Given a recurrence relation, solve it using the substitution method:

# a. T(n) = T(n-1) + c
'''
Let Base Case : T(1) = c

T(n) = T(n-1) + c .....(1)
T(n-1) = T(n-2) + c .....(2)
T(n-2) = T(n-3) + c .....(3)

Add (1), (2) and (3)

T(n) = T(n-3) + 3c  Three times
.
.
.
K times
T(n) = T(n-k) + kc
let n-k = 1 or k = n-1
T(n) = T(1) + (n-1)c
T(n) = c + (n-1)c
T(n) = n*c  where c is constant

Time Complexity = O(n)

'''
```

```python
# b. T(n) = 2T(n/2) + n
'''
T(n) = 2T(n/2) + n
Let Base Case : T(1) = 1

T(n) = 2T(n/2) + n                (1)
T(n/2) = 2T(n/2^2) + n/2          (2)
T(n/2^2) = 2T(n/2^3) + n/2^2      (3)
T(n/2^3) = 2T(n/2^4) + n/2^3      (4)

add Equation (1) + 2*(2) + (2^2)*(3) + (2^3)*(4)
T(n) = (2^4)T(n/2^4) + n + n + n+ n  4th times
.
.
.
Kth times:
T(n) = (2^k)T(n/2^k) + k*n
Let n/2^k = 1   or   2^k = n
solving them we get  k = log₂n

T(n) = nT(1) + n*log₂n   : T(1) = 1
T(n) = n + n*log₂n

Time Complexity : O(n*log₂n)
'''
```

```python
# c. T(n) = 2T(n/2) + c
'''
T(n) = 2T(n/2) + c
Let Base Case : T(1) = 1

T(n) = 2T(n/2) + c              (1)
T(n/2) = 2T(n/2^2) + c          (2)
T(n/2^2) = 2T(n/2^3) + c        (3)
T(n/2^3) = 2T(n/2^4) + c        (4)

add Equation (1) + 2*(2) + (2^2)*(3) + (2^3)*(4)
T(n) = (2^4)T(n/2^4) + c + 2c + 2^2c + 2^3c  4th times
.
.
.
Kth times:
T(n) = (2^k)T(n/2^k) + c + 2c + 2^2c + 2^3c + .....+ 2^(k-1)c
Let n/2^k = 1   or   2^k = n
solving them we get  k = log₂n
T(n) = nT(1) + 2^k - 1
T(n) = n*1 + n -1
T(n) = 2n - 1

Time Complexity : O(n)
'''
```

```python
# d. T(n) = T(n/2) + c
'''
T(n) = T(n/2) + c
Let Base Case : T(1) = 1

T(n) = T(n/2) + c              (1)
T(n/2) = T(n/2^2) + c          (2)
T(n/2^2) = T(n/2^3) + c        (3)
T(n/2^3) = T(n/2^4) + c        (4)

add Equation (1) + (2) + (3) + (4)
T(n) = T(n/2^4) + c + c + c + c  4th times
.
.
.
Kth times:
T(n) = T(n/2^k) + k*c
Let n/2^k = 1   or   2^k = n
solving them we get  k = log₂n
T(n) = T(1) + log₂n*c
T(n) = 1 + c*log₂n

Time Complexity : O(log₂n)
'''
```

```python
# 3. Given a recurrence relation, solve it using the recursive tree approach:

# a. T(n) = 2T(n-1) + 1
'''
To solve the recurrence relation T(n) = 2T(n-1) + 1 using the recursive tree approach, we can visualize the recursive calls in a tree structure.
The recurrence relation suggests that each level of the recursion involves two recursive calls with the argument n-1n−1, and there is an
additional constant term of 1 at each level. 
Each level of the tree contributes a constant term of 1. The total number of levels is nn, as we are decrementing nn by 1 at each level until
we reach the base case.
T(n) = 2T(n-1) + 1                        1st Level
T(n) = 2[2T(n-2) + 1] + 1                 2nd Level
T(n) = 2[2[2T(n-3) + 1] + 1] + 1           3rd Level
T(n) = 2[2[2[2T(n-4) + 1] + 1] + 1] + 1    4th Level
T(n) = 2^4T(n-4) + 1 + 2 + 2^2 + 2^3       4th Level
.
.
.
.
.
T(n) = 2^kT(n-k) + 1 + 2 + 2^2 + 2^3 + .......+ 2^(k-1)
Let n-k = 0 or k = n
T(n) = T(0) + 1 + 2 + 2^2 + 2^3 + .......+ 2^(n-1)

Let T(0) = 1
T(n) = 1 + Sum of G.P Series Common ratio = 2
T(n) = 1 + 2^n - 1
T(n) = 2^n

Time complexity : O(2^n)[Exponential]
'''
```
